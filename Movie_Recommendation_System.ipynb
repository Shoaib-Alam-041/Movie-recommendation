**Importing the Dependencies**

```python
import pandas as pd
import numpy as np
import difflib #Text comparison
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

**Data Collection and Pre-Processing**

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#Selecting Relevant features
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
movies_data.isna().sum()
```

```python
#Filling missing Values with empty String
for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(combined_features)
```

```python
# Converting the Text to Features vector
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

**Cosine Similarity**

1.   It shows the vectors similarity

2.   1- Text is same

3.  0- Text is compeletly different



```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
user_movie = input("PLease Enter a Movie Name: ")
```

```python
list_of_all_movies = movies_data['title'].tolist()
list_of_all_movies
```

```python
find_close_match = difflib.get_close_matches(user_movie,list_of_all_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# Finding the index of the move wit title
movie_index = movies_data[movies_data.title == close_match ]['index'].values[0]
print(movie_index)
```

```python
# Getting a list of similar movies
similarity_score = list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
# Soting the movies
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# Print the name of similar movie based on index
print("Movies suggested for you: \n")

i=1

for movie in sorted_similar_movies:
  index = movie[0]
  tital_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<10):
    print(i, '.', tital_from_index)
    i+=1
```

**Movie Recommendation System**

```python
user_movie = input("PLease Enter a Movie Name: ")
list_of_all_movies = movies_data['title'].tolist()
find_close_match = difflib.get_close_matches(user_movie,list_of_all_movies)
close_match = find_close_match[0]
movie_index = movies_data[movies_data.title == close_match ]['index'].values[0]
similarity_score = list(enumerate(similarity[movie_index]))
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print("Movies suggested for you: \n")

i=1

for movie in sorted_similar_movies:
  index = movie[0]
  tital_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<10):
    print(i, '.', tital_from_index)
    i+=1
```
